```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2023)
a = np.random.randint(1,100,20)
a
```

```python
plt.plot (a)
```

```python
lakes = []
current_depth = a[0]
current_lake = [a[0]]

for i in a[1:]:
    if i <= current_depth:
        current_lake.append(i)
    else:
        current_lake.append(i)
        lakes.append(current_lake)
        current_lake = [i]
        current_depth = i

lakes.append(current_lake)

lakes
```

```python
deepest_lake = max(lakes, key=lambda lake: max(lake) - min(lake))
deepest_lake
```

```python
colors = ['blue' if array != deepest_lake else 'red' for array in lakes]

position = 0

for i in range(len(lakes)):
    x = [j + position for j in range(len(lakes[i]))]
    y = lakes[i]
    plt.plot(x, y, c=colors[i], linestyle='-', label=f'Masiv {i+1}')
    position += len(lakes[i]) - 1
```
